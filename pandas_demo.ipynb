#### pd.read_excel
pd.read_excel(io, sheetname=0,header=0,skiprows=None,index_col=None,names=None,
                arse_cols=None,date_parser=None,na_values=None,thousands=None, 
                convert_float=True,has_index_names=None,converters=None,dtype=None,
                true_values=None,false_values=None,engine=None,squeeze=False,**kwds)  
常用参数解析：
* io ：excel 路径；
* sheetname：默认是sheetname为0，返回多表使用sheetname=[0,1]，若sheetname=None是返回全表 。注意：int/string返回的是dataframe，而none和list返回的是dict of dataframe。
* header ：指定作为列名的行，默认0，即取第一行，数据为列名行以下的数据；若数据不含列名，则设定 header = None；
* skiprows：省略指定行数的数据
* skip_footer：省略从尾部数的行数据
* index_col ：指定列为索引列，也可以使用 u’string’
* names：指定列的名字，传入一个list数据

```python
# 分组统计
df7 = pd.DataFrame(data=np.array([0,1,2,0,3,4,0,6,3,1,4,3,1,5,7,2,3,5,2,7,8]).reshape(-1,3), columns=[*"ABC"])
df7
df7.groupby('A')
df7.groupby('A')['B']
df7.groupby('A')['B'].mean()
# 计算多个函数的值用函数aggregate，将函数名以列表的形式枚举在参数中
df7.groupby('A')['B'].aggregate(["mean","sum","max","median",np.min])
# unstack 转换为二级索引
a = df7.groupby('A')['B'].aggregate(["mean","sum","max","median",np.min]).unstack()
a
a.index
# 如果将第3个位置上的函数替换为自定义函数，则用方法apply()
def myfunc(x):
    x['B'] /= x['B'].sum()
    return x
def myfunc1(x):
    x+=1
    return x
df7.groupby('A').apply(myfunc)
df7.groupby('A')['B'].apply(myfunc1)
```

# pandas绘图函数
Series和DataFrame都有一个用于生成各类图表的plot方法。默认情况下，它们所生成的是线形图

matplotlib来绘制
格式： s.plot(kind=‘line’, ax=None, figsize=None, use_index=True, title=None, grid=None, legend=False, style=None, logx=False, logy=False, loglog=False, xticks=None, yticks=None, xlim=None, ylim=None, rot=None, fontsize=None, colormap=None, table=False, yerr=None, xerr=None, label=None, secondary_y=False, **kwds)  

参数kind用来指定绘制图像类型：

参数|类型
:---:|:---:
‘line’|line plot (default)
‘bar’| vertical bar plot
‘barh’| horizontal bar plot
‘hist’| histogram
‘box’|  boxplot
‘kde’| Kernel Density Estimation plot
‘density’| same as ‘kde’
‘area’| area plot
‘pie’| pie plot
